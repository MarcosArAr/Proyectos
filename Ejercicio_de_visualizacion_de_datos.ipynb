## VISUALIZACION

```python
df.groupby(["Race","Year","Gender"])[["Survival Rate"]].mean().round(3).unstack(level="Year")
```

```python
df_gendertotal = df[df["Gender"] != "total"]
df_graf = df_gendertotal[df_gendertotal["Cancer Type"] != "All cancers"]
df_aux = df_graf[df_graf["Race"] != "All races"]
df_aux
```

```python
df_aux_females = df_aux[df_aux["Gender"] == "females"]
df_aux_females
```

```python
df_aux_males = df_aux[df_aux["Gender"] == "males"]
df_aux_males
```

```python
df_auxfem = df_aux_females.groupby(["Year"])[["Survival Rate"]].mean().round(3).sort_values(by="Survival Rate")
df_auxfem
```

```python
plt.plot(df_auxfem.index , df_auxfem["Survival Rate"].values, color="pink")
```

```python
df_auxmal = df_aux_males.groupby(["Year"])[["Survival Rate"]].mean().round(3).sort_values(by="Survival Rate")
df_auxmal
```

```python
plt.plot(df_auxmal.index , df_auxmal["Survival Rate"].values, color="blue")
```

```python
plt.figure(dpi=80)
plt.plot(df_auxmal.index , df_auxmal["Survival Rate"].values, color="blue", label="Hombres")
plt.plot(df_auxfem.index , df_auxfem["Survival Rate"].values, color="pink", label="Mujeres")
plt.xlabel("AÃ±o")
plt.ylabel("Promedio de la tasa de supervivencia")
plt.title("Diferencia de la tasa de supervivencia entre hombres y mujeres segÃºn aÃ±o")
plt.legend()
plt.show()
```
